# Exploratory Data Analysis(EDA)

## For EDA these steps were performed
1. Concatenate all coins into a single table.
2. Rename Pairs to Symbol. 
3. Rename Source to Exchange (abbreviated as 'ex').
4. Create the columns "Base" and "Quote" from "Symbol" so BTC/USD is base=BTC quote=USD.

```python
#importing libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

### 1. Concatenate all coins into a single table.

```python
# reading each coin csv files with pandas
df1 = pd.read_csv("Bitcoin.csv", index_col=0)
df2 = pd.read_csv("Etherium.csv", index_col=0)
df3 = pd.read_csv("Binance_coin.csv", index_col=0)
df4 = pd.read_csv("Cardano.csv", index_col=0)
df5 = pd.read_csv("Tether.csv", index_col=0)
df6 = pd.read_csv("XRP.csv", index_col=0)
df7 = pd.read_csv("Solana.csv", index_col=0)
df8 = pd.read_csv("Polkadot.csv", index_col=0)
df9 = pd.read_csv("USD_coin.csv", index_col=0)
df10 = pd.read_csv("Dogecoin.csv", index_col=0)
df11 = pd.read_csv("Terra.csv", index_col=0)
df12 = pd.read_csv("Uniswap.csv", index_col=0)
df13 = pd.read_csv("Avalanche.csv", index_col=0)
df14 = pd.read_csv("Binance_USD.csv", index_col=0)
df15 = pd.read_csv("Chainlink.csv", index_col=0)
df16 = pd.read_csv("Litecoin.csv", index_col=0)
df17 = pd.read_csv("Algorand.csv", index_col=0)
df18 = pd.read_csv("Bitcoin_Cash.csv", index_col=0)
df19 = pd.read_csv("Wrapped_Bitcoin.csv", index_col=0)
df20 = pd.read_csv("SHIBA_INU.csv", index_col=0)


# display(df1.head(2))
# display(df2.head(2))
# display(df3.head(2))
# display(df4.head(2))
# display(df5.head(2))
# display(df6.head(2))
# display(df7.head(2))
# display(df8.head(2))
# display(df9.head(2))
# display(df10.head(2))
```

```python
# concatenating all dataframes vertically into a single dataframe "all_coins"
all_coins = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15, df16, df17, df18, df19, df20], axis=0)

#creating a new index for the new dataframe
all_coins = all_coins.reset_index()

#dropping old index
all_coins.drop('index', axis=1, inplace=True)

#viewing the top and bottom rows of our result
display(all_coins.head())
display(all_coins.tail())
```

### 2. Renaming columns

```python
# renaming the columns appropriately
all_coins.rename(columns = {'source':'ex','pairs':'symbol', 'plus_2_Depth':'+2% depth', 'minus_2_Depth':'-2% depth', 'volume_perc':'volume%'}, inplace = True)

#viewing columns to confirm change
display(all_coins.columns)

display(all_coins.head())
```

### 4. Create the columns 'Base' and 'Quote' from 'Symbol' so BTC/USD is base=BTC and quote=USD.

```python
#splitting the 'symbol' column into two new columns 'base' and 'quote'
all_coins[['base','quote']] = all_coins.symbol.str.split('/', expand=True)

#viewing our resulting dataframe
display(all_coins.columns)
display(all_coins.head())
```

# Second section: Visualize

## In this section these steps were performed
1. Create a table 
> where, 
>- for each tradingpair(symbol), its top most liquid exchanges are measured by the column "Volume"
>- columns: symbol, ex, volume, volume%, +2% depth, -2% depth
>- title: volume on the selected markets yyyy-mm-dd (cmc).
>- replace yyyy-mm-dd with the date grabbed the data was grabbed

2. Display 'volume' column in scientific notation(float) in the table if possible using two decimals e.g. 1.23E6 since we want to look at size, not its precision.

```python
# creating a new dataframe with the columns required
new_df = all_coins[['symbol', 'ex', 'volume', 'volume%', '+2% depth', '-2% depth', 'quote']]

new_df.head()
```

```python
# removing the '$' and ',' in volumn column 
new_df['volume'] = new_df['volume'].apply(lambda x: x.replace('$',''))
new_df['volume'] = new_df['volume'].apply(lambda x: x.replace(',',''))

#Display 'volume' in scientific notation(float)
new_df['volume'] = new_df['volume'].astype(float)
```

```python
#viewing our end results
new_df[['volume']]
```

```python
# showing top liquid exchanges as measured by the column "Volume" for each tradingpair
# assigning the resulting table as 'volume_on_the_selected_markets_2021_10_07_cmc'

volume_on_the_selected_markets_2021_10_07_cmc= new_df.groupby(['symbol', 'ex'])[['volume']].sum()

# saving it locally as a csv file
volume_on_the_selected_markets_2021_10_07_cmc.to_csv('volume_on_the_selected_markets_2021_10_07_cmc.csv')
```

# Section 3: Plot

## In this section these steps were carried out: 
### Plot
>title: volume on the selected markets yyyy-mm-dd (cmc)

>a stacked bar chart of corresponding table in section 2.

>x axis: 'quote' column.

>y axis: sum of 'volume' column

>color by: 'exchange' column

>legend: name of 'exchange'

> graph all quotes including /USD and /USDT.

- Save one .png for each asset

- Create one plot for each asset 

- put all asset inside a 5 by 4 matrix.

```python
#importing library
import matplotlib.pyplot as plt
```

```python
# assigning corresponding table to variable graph for this section
graph = new_df.groupby(['symbol', 'ex'])['volume'].sum().unstack()
```

```python
# in order to plot we need to select each asset
#viewing the index of assets
a = -1
for x in graph.index:
    a+=1
    print('{}, {}'.format(a, x))
    
```
